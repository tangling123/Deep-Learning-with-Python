## 2.3.1 逐元素运算

relu和加法都是逐元素运算  
relu(x) = max(x, 0)  
### Python中assert断言，用于判断是否满足条件  
### 条件为false,则直接返回错误

```python
def native_relu(x):
    assert len(x.shape) == 2
    
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            x[i][j] = max(x[i][j], 0)
    return x
```

```python
def native_add(x, y):
    assert len(x.shape) == 2
    assert x.shape == y.shape
    
    for i in len(x.shape[0]):
        for j in len(x.shape[1]):
            x[i][j] += y[i][j]
    return x
```

## 2.3.2 广播
两个张量大小不一致时，会将较小的张量进行广播  
(1)向较小的张量添加轴，使其ndim与较大的张量相同  
(2)将较小的张量沿着新轴重复，使其与较大的张量相同

将矩阵和向量相加

```python
def native_add_matrix_and_vector(x,y):
    assert len(x.shape) == 2
    assert len(y.shape) == 1
    assert x.shape[1] == y.shape[0]
    
    x = x.copy()
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            x[i][j] += y[j]
    return x
```

```python
import numpy as np

x = np.random.random((64, 3, 32, 10))
y = np.random.random((32, 10))

z = np.maximum(x, y)
```

```python
print(f" x.shape: {x.shape}\n y.shape:{y.shape}\n z.shape:{z.shape}")
print(z)
```

## 2.3.3 张量点积

1、向量之间的点积  
两个向量做点积，对应位置上元素相乘并相加，最后得到一个数值

```python
def native_vector_dot(x, y):
    assert len(x.shape) == 1
    assert len(y.shape) == 1
    assert x.shape[0] == y.shape[0]
    
    int res = 0
    for i in range(x.shape[0]):
        res += x[i] * y[i]
    return res
```
