## 4.1 机器学习的四个分支
前面学到的二分类问题、多分类问题、回归问题和标量回归，这三者都是监督学习的例子  
其目标是学习训练输入和训练目标之间的关系  

### 4.1.1监督学习
监督学习是目前最常见的机器学习类型  
给定一组样本（通常由人工标注），他可以学会将输入数据映射到已知目标(标注)  
监督学习主要分为分类和回归，但还有更多的变体，主要包括以下几种：  
#### (1)序列生成(sequence generation)。
给定一张图像，预测描述图像的文字，序列生成有时可以被重新表示一系列分类问题，比如反复预测序列中的单词或标记。  
#### (2)语法树检测(syntax tree prediction)
给定一个句子，预测其分解生成的语法树  
#### (3)目标检测(object detection)
给定一张图像，在图中特定目标的周围画一个边界框。  
这个问题也可以表示为分类问题（给定多个候选框边界，对每个框内的目标进行分类）  
或分类与联合回归问题（用向量回归来预测边界框的坐标）
#### (4)图像分割(image segmentation)
给定一张图像，在特定物体上画一个像素级的掩膜(mask)

### 4.1.2 无监督学习
无监督学习是指在没有目标的情况下寻找输入数据的有趣变换，其目的在于数据可视化、数据压缩、数据去噪或更好的理解数据中的相关性。  
无监督学习是数据分析的必备技能，在解决监督学习问题之前，为了更好的理解数据集，他通常是一个必要步骤。  
#### 降维(dimensionality reduction)和聚类(clusing)都是众所周知的无监督学习方法

### 4.1.3 自监督学习
自监督学习是监督学习的一个特例，他与众不同，值得单独归一类。  
自监督学习是没有人工标注的标签的监督学习，可以看做是没有人类参与的监督学习，但标签仍然存在（因为总要有什么东西来监督学习过程），但他们是从输入数据中生成的，通常是使用启发式的算法生成的。  

#### 自编码器(autoencoder)，是著名的自监督学习的例子，其生成的目标就是未经修改的输入  
#### 时序监督学习(temporally supervised learning)（用未来的输入数据作为监督）  
给定视频过去的帧来预测下一帧或给定文本前面的词来预测下一个词  

#### notes : 无监督学习、自监督学习和无监督学习之间的区别有时候很模糊，这三个类别更像是没有明确边界的连续体，本书的重点在于监督学习

### 4.1.4 强化学习  
#### 在强化学习中，智能体(agent)接受其环境信息，并学会选择使用某种奖励最大化的行动  
强化学习主要由智能体（Agent）、环境（Environment）、状态（State）、动作（Action）、奖励（Reward）组成。智能体执行了某个动作后，环境将会转换到一个新的状态，对于该新的状态环境会给出奖励信号（正奖励或者负奖励）。随后，智能体根据新的状态和环境反馈的奖励，按照一定的策略执行新的动作。上述过程为智能体和环境通过状态、动作、奖励进行交互的方式。
