### 4.5.1 定义问题、收集数据集
只有明确输入、输出和所使用的数据集，才能进入下一阶段  
* 假设输出是可以根据输入来进行预测的  
* 假设可用数据包含足够多的信息，足以学习输入和输出之间的关系  
机器学习只能用来记忆训练数据中存在的模式，在过去的数据上来预测未来（前提是未来的规律和过去相同）

### 4.5.2 选择衡量成功的指标
准确率(precision)、召回率(recall)、F1-score  
* 平衡分类问题，精度和接收者操作曲线下面积(area under receiver operating characteristic curve, ROC AUC)是常用的指标
* 对于类别不平衡的问题，可以使用准确率和召回率  
* 对于排序或多标签分类，可以使用平均准确均值(mean average precision)
* 自定义成功的指标

### 4.5.3 确定评估方法
* 留出验证集  
* K折交叉验证
* 重复的K折验证

### 4.5.4 准备数据(数据的预处理)
* 在数据输入神经网络之前应该先转换为张量  
* 这些张量的取值通常应该缩放为较小的值，比如在[0, 1]之间
* 如果不同特征有不同的取值范围，需要将数据进行标准化
* 特征工程

### 4.5.5 开发比基准更好的模型
使用随机的模型作为基准(dumb baseline)   


### 4.5.6 扩大模型的规模：开发过拟合模型
机器学习中无处不在对立的是优化和泛化的对立，理想的模型师建立在欠拟合好过拟合的界线上，在容量不足和容量过大的界线上，为了找到这条界线，你必须穿过它  
开发一个过拟合的模型：  
* 添加更多的层
* 让每一层变大
* 训练更多的轮次

### 4.5.7 模型正则化和调节超参数
* 添加dropout
* 尝试不同的架构，增加或减少层数
* 添加L1或L2正则化
* 尝试不同的超参数
* 反复做特征工程：添加新的特征或删除没有信息量的特征
一旦开发出满意的模型，就可以将训练数据和验证数据训练到最终的模型，然后在测试集上进行评估
