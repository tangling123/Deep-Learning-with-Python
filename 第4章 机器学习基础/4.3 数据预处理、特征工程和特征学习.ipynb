## 4.3 数据预处理、特征工程和特征学习
在建立神经网络之前，首先应该解决一个重要问题：将数据输入神经网络之前，如何准备数据数据和目标  

### 4.3.1 神经网络的数据预处理
神经网络预处理的目的是使原始数据更适用于神经网络处理，包括向量化、标准化、处理缺失值和特征提取  
#### 1、向量化
神经网络的所有输入和目标都必须是浮点数张量(在特定情况下可以是整数张量)  
**数据向量化：不能将整数序列直接输入神经网络，需要将列表转换为张量**  
#### 转换的方法有两种:  
#### (1)填充列表，使其具有相同的长度，再将列表转换成形状为(samples, word_index)的整数张量  
#### 然后网络的第一层使用能处理这种张量的层(Embedding层)    

#### (2)对列表进行one_hot编码，将其转换为0和1组成的向量  
#### 然后网络的第一层可以使用Dense层，它能处理浮点数向量数据  

#### 2、值标准化
特征有不同的取值范围时，在数据输入网络之前，需要对每个特征分别作标准化  
将取值相对较大的数据或异质数据(比如一个特征在0-1范围，另一个特征在100-200),输入到神经网络当中是不安全的。这么做可能导致较大的梯度更新，进而导致网络无法收敛，为了让网络的学习变得容易，输入的数据应该具有以下特征:
#### (1)取值较小
大部分取值都应该在0-1之间
#### (2)同质性
所有的特征都应该在大致相同的范围内  
  
标准化方法： 
(1)将每个特征分别标准化，使其平均值为0  
(2)将每个特征分别标准化，使其标准差为1  
mean = x.mean(axis = 0) #假设x是一个形状为(samples, features)的二维矩阵  
std = x.std(axis = 0)  
x -= mean
x /= std

#### 3、处理缺失值
一般来说，对于神经网络，将缺失值设置为0是安全的，只要0不是一个有意义的值  
网络能够从数据学到0意味着缺失数据，并且会忽略这个值  

### 4.3.2 特征工程
特征工程(feature engineering)是指将数据输入到模型之前，利用自己关于数据和机器学习算法的知识对数据进行硬编码的变换，以改善模型的效果  
(1)机器学习 --> (2)(x,y)之间的坐标 --> (3)两个指针的角度

特征工程的本质：用更简单的方式，从而使问题变得更容易  
#### Advantages:
1、良好的特征可以让你用更少的资源更优雅的解决问题  
2、良好的特征可以让你用更少的数据解决问题
