## 4.4 过拟合与欠拟合
随着模型的不断训练，模型在训练集上的表现越来越好，但在验证集上的表现却不在变好甚至开始下降，这时候就出现了模型的过拟合  
* 机器学习的根本问题就是优化和泛化之间的对立 
* 优化(optimization)：是指调节模型以在训练集上得到最佳性能(即机器学习中的学习)  
* 泛化(generalization)：是指训练好的模型在前所未见的数据上的性能好坏  

* 为了防止模型从训练数据中学到错误或无关紧要的模式，最优解决办法是获取更多的训练数据 
* 次优解决办法是调节模型允许存储的信息量，或对模型允许存储的信息加以约束 
* 降低网络过拟合的方法叫正则化(regularization) 

### 4.4.1 减小网络大小
防止过拟合的最简单的方法就是减小模型的大小，即减少模型中可以学习参数的个数(这由层数和每层单元个数决定)  
#### 模型的容量(capacity)：模型中可以学习参数的个数
#### 深度学习模型通常都很擅长拟合数据，但真正的挑战在于泛化，而不是拟合
在电影评论分类的网络上修改网络的大小

```python
#1、从模型中加载数据集
from keras.datasets import imdb

(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words = 10000)

#2、数据的预处理
import numpy as np

def vectorize_sequences(sequences, dimension = 10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results

train_data = vectorize_sequences(train_data)
train_labels = np.asarray(train_labels).astype('float32')

#3、留出验证集
x_val = train_data[ : 10000]
y_val = train_labels[ : 10000]

x_train = train_data[10000 : ]
y_train = train_labels[10000 : ]

#4、构建网络
from keras import models, layers

def build_network(size):
    network = models.Sequential()
    network.add(layers.Dense(size, activation = 'relu', input_shape = (10000, )))
    network.add(layers.Dense(size, activation = 'relu'))
    network.add(layers.Dense(1, activation = 'sigmoid'))
    
    network.compile(optimizer = 'rmsprop',
                    loss = 'binary_crossentropy',
                    metrics = 'accuracy')
    
    return network

#5、训练模型
network_16 = build_network(16)
network_4 = build_network(4)
network_32 = build_network(32)

history_16 = network_16.fit(x_train, y_train, epochs = 20, batch_size = 512, validation_data = (x_val, y_val))
history_4 = network_4.fit(x_train, y_train, epochs = 20, batch_size = 512, validation_data = (x_val, y_val))
history_32 = network_32.fit(x_train, y_train, epochs = 20, batch_size = 512, validation_data = (x_val, y_val))

#6、绘制验证结果
import matplotlib.pyplot as plt

loss = history_16.history['val_loss']
loss_4 = history_4.history['val_loss']
loss_32 = history_32.history['val_loss']
epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'blue', label = 'size = 16')
plt.plot(epochs, loss_4, 'grey', label = 'size = 4')
plt.plot(epochs, loss_32, 'orange', label = 'size = 32')
plt.title('the impact of network size')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.legend()

plt.show()
```

### 4.4.2 添加权证正则化
奥卡姆剃刀(Ocaam's razor)原理:如果一件事情有两种解释，那么最可能正确的解释就是最简单的那个，即假设更少的那个  
这个原理也适用于神经网络学到的模型，简单的网络比复杂度网络更不容易发生过拟合  
* 简单模型是指参数更少的模型

#### 权重正则化(weight regularization)：强制让模型的权重只能取较小的值，从而降低模型的复杂度，使得权重分布更加规则(regular)，其实现方法就是向网络损失函数中添加与较大权重值相关的成本(cost)

* L1正则化(L1 regularization)：添加的成本与权重系数绝对值成正比  
* L2正则化(L2 regularization)：添加到成本与权重系数的平方成正比，神经网络的L2正则化也叫权重衰减(weight decay)
在Keras中，添加权重正则化的方法是向层传递权重正则化实例(weight regularizer instance)作为关键字参数

```python
#向模型添加L2正则化

from tensorflow.keras import regularizers
from keras import models, layers

#建立一个L2正则化的网络
network = models.Sequential()
network.add(layers.Dense(16, kernel_regularizer = regularizers.l2(0.001), 
            activation = 'relu', input_shape = (10000, )))
network.add(layers.Dense(16, kernel_regularizer = regularizers.l2(0.001)))
network.add(layers.Dense(1, activation = 'sigmoid'))
network.compile(optimizer = 'rmsprop',
                loss = 'binary_crossentropy',
                metrics = 'accuracy')


network_original = build_network(16)

#训练模型
history = network.fit(x_train, y_train, epochs = 20, 
                      batch_size = 512, validation_data = (x_val, y_val))

history_original = network_original.fit(x_train, y_train, epochs = 20, 
                                        batch_size = 512, validation_data = (x_val, y_val))


#6、绘制验证结果
import matplotlib.pyplot as plt

loss = history.history['val_loss']
loss_original = history_original.history['val_loss']
epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'blue', label = 'L2-regularized model')
plt.plot(epochs, loss_original, 'orange', label = 'Original model')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.legend()

plt.show()
```

### 4.4.3 添加Dropout正则化
dropout正则化是神经网络最有效也是最常用的正则化之一  
对某一层使用dropout正则化，就是在训练过程中随机将该层的一些输出特征舍弃(设置为0)  
dropout比例(dropout rate)是别设为0的特征所占的比例，通常在0.2-0.5之间  
测试时，没有单元被舍弃，而该层的输出值需要按照dropout比率缩小，因为这时比训练时有更多的单元被激活，需要加以平衡 

```python
#向模型添加droput正则化

from tensorflow.keras import regularizers
from keras import models, layers

#建立一个L2正则化的网络
network = models.Sequential()
network.add(layers.Dense(16, activation = 'relu', input_shape = (10000, )))
network.add(layers.Dropout(0.5))
network.add(layers.Dense(16, activation = 'relu'))
network.add(layers.Dropout(0.5))
network.add(layers.Dense(1, activation = 'sigmoid'))
network.compile(optimizer = 'rmsprop',
                loss = 'binary_crossentropy',
                metrics = 'accuracy')


network_original = build_network(16)

#训练模型
history = network.fit(x_train, y_train, epochs = 20, 
                      batch_size = 512, validation_data = (x_val, y_val))

history_original = network_original.fit(x_train, y_train, epochs = 20, 
                                        batch_size = 512, validation_data = (x_val, y_val))


#绘制验证结果
import matplotlib.pyplot as plt

loss = history.history['val_loss']
loss_original = history_original.history['val_loss']
epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'blue', label = 'Dropout-regularized model')
plt.plot(epochs, loss_original, 'orange', label = 'Original model')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.legend()

plt.show()
```

防止神经网络过拟合的方法：  
* 获取更多的训练数据  
* 减小网络容量  
* 添加权重正则化  
* 添加dropout正则化
